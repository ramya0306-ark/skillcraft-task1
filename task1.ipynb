```python
top_10_population_2022 = df.sort_values(by='2022_population', ascending=False).head(10)
display(top_10_population_2022[['country/territory', '2022_population']])
```

BAR-BLOT

**TOP 10 COUNTRIES OF TOTAL POPULATION IN YEAR 2022 AND 2016**

```python
import matplotlib.pyplot as plt
categories = top_10_population_2022['country/territory']
values = top_10_population_2022['2022_population']

plt.figure(figsize=(13, 6))
plt.bar(x=categories, height=values, color='skyblue')
plt.title('Top 10 Countries by Population in 2022')
plt.xlabel('Country/Territory')
plt.ylabel('Population (2022)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

```python
total_population_sorted1 = df.sort_values(by='2022_population', ascending=True)
total_bottom_ten_countries = total_population_sorted1.head(10)
print("Bottom ten countries of total population\n")
display(total_bottom_ten_countries[['country/territory', '2022_population']])
```

```python
import matplotlib.pyplot as plt
categories = total_bottom_ten_countries['country/territory']
values = total_bottom_ten_countries['2022_population']

plt.figure(figsize=(12, 6))
plt.bar(x=categories, height=values, color='purple')
plt.title('Bottom 10 Countries by Population in 2022')
plt.xlabel('Country/Territory')
plt.ylabel('Population (2022)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```

**GROWTH RATE**

```python
growth_rate_data = df['growth_rate']
display(growth_rate_data.head())
display(growth_rate_data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='growth_rate', kde=True)
plt.title('Distribution of Growth Rate')
plt.xlabel('Growth Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_population_2022, x='country/territory', y='world_population_percentage', palette='viridis')
plt.title('World Population Percentage for Top 10 Countries (2022)')
plt.xlabel('Country/Territory')
plt.ylabel('World Population Percentage')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='density_(per_km²)', y='growth_rate')
plt.title('Population Density vs. Growth Rate')
plt.xlabel('Population Density (per km²)')
plt.ylabel('Growth Rate')
plt.show()
```
